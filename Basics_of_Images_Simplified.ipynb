# Resizing

# Grayscale conversion

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(gray_image,cmap="gray")
plt.show()
```

# Blurring

# Edge Detection

Highlight the edges in an image using Canny edge detector

```python
img_canny = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(img_canny, threshold1=10, threshold2=200)#, cv2.COLOR_RGB2GRAY)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(edges,cmap="gray")
plt.show()
```

# Thresholding

Converts a grayscale image into a binary image

Pixels above a certain threshold are set to 255 and below to 0.

```python
_, binary_image = cv2.threshold(gray_image,70,255,cv2.THRESH_BINARY) #multiplied by .
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(binary_image,cmap="gray")
plt.show()
```

# Morphological Operations

- Dilation and Erosion is used to process the structures of objects in an image

- Dilation: Expands the white regions in a binary image

- Erosion shrinks the white images in a binary image

```python
############## Dilation #########################
#kernel
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5, 5), np.uint8)
#dilated
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(dilated_image,cmap="gray")
plt.show()
```

```python
############## Closing ######################
#kernel
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5, 5), np.uint8)
#dilated
eroded_image = cv2.erode(dilated_image, kernel, iterations=1)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(eroded_image,cmap="gray")
plt.show()
```

```python
############## Erosion ######################
#kernel
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5, 5), np.uint8)
eroded_image = cv2.erode(binary_image, kernel, iterations=1)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(eroded_image,cmap="gray")
plt.show()
```

```python
############## Opening #########################
#kernel
# Taking a matrix of size 5 as the kernel
kernel = np.ones((6, 6), np.uint8)
#dilated
dilated_image = cv2.dilate(eroded_image, kernel, iterations=1)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(dilated_image,cmap="gray")
plt.show()
```

# Image Rotation and Translation

```python
######################### Rotation ##################################################################

rows, cols = img.shape[:2]
matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 65, 1)
rotated_image = cv2.warpAffine(img, matrix, (cols, rows))
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(rotated_image,cmap="gray")
plt.show()
```

```python
######################### Translation ##################################################################
matrix = np.float32([[1, 0, 50], [0, 1, 50]])  # Move 50 pixels right and down
translated_image = cv2.warpAffine(img, matrix, (cols, rows))
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(translated_image,cmap="gray")
plt.show()
```

# Histogram Equalization

- Improves constast of an image by distributing the intensity values more evenly

```python
gray_image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(gray_image,cmap="gray")
plt.show()
```

```python
equalized_image = cv2.equalizeHist(gray_image)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(equalized_image,cmap="gray")
plt.show()
```
